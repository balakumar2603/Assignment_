```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independant=dataset[['age','bmi','children','sex_male','smoker_yes']]
independant
```

```python
dependant=dataset[['charges']]
dependant
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['mse','mae'],
                'max_features': ['auto','sqrt','log2'],
                    'n_estimators':[10,100]}
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose=3, n_jobs=-1)


grid.fit(X_train,y_train)
```

```python
re=grid.cv_results_

```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
Future_predictions=grid.predict([[age,bmi,children,sex,smoker]])
Future_predictions
```
