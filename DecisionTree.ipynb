```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
independant=dataset[['age','bmi','children','sex_male','smoker_yes']]
independant
```

```python
dependant=dataset[["charges"]]
dependant
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='mse',max_features='log2',splitter='random')
regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_predict= regressor.predict(X_test)
y_predict
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_predict)
r_score
```

```python
import pickle
filename="finalized_model_decision_tree.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
load_model=pickle.load(open(filename,'rb'))
```
